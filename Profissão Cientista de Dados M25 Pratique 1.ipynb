2. Dada uma base de dados com uma variável resposta $y$ e um conjunto de variáveis explicativas. Considere uma estrutura de um modelo de regressão. Explique com suas palavras por que não é possível obter parâmetros que forneçam um erro quadrático médio (EQM) menor que o obtido com estimadores de mínimos quadrados.

Porque o estimador de mínimos quadrados é definido exatamente como o conjunto de parâmetros que minimiza a soma dos erros ao quadrado (e, consequentemente, o EQM) nos dados usados no ajuste. Para a regressão linear, essa função de custo é quadrática e convexa em relação aos parâmetros, então existe um mínimo global. Assim, qualquer outro conjunto de parâmetros dentro da mesma estrutura de modelo não consegue produzir um EQM menor na base de treino; no máximo empata (no mesmo mínimo), ou piora. Para obter EQM menor, seria necessário alterar a estrutura do modelo, as hipóteses, as variáveis, ou avaliar em outro conjunto de dados.
