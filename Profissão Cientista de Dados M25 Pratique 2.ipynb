### 1) Integral numérica (método dos retângulos)

Vamos aproximar a área sob a curva de \(f(x)=x^3\) no intervalo \([1,2]\) usando retângulos.
Escolhi o método do **ponto médio**, que costuma dar uma aproximação melhor.

- Intervalo: \([1,2]\)
- Número de retângulos: \(n = 10000\)
- Largura de cada retângulo: \(\Delta x = \frac{2-1}{n}\)
- Pontos médios: \(x_i = 1 + \left(i+\frac{1}{2}\right)\Delta x\)
- Área aproximada: \(\sum f(x_i)\Delta x\)


```python
import numpy as np

# Função e intervalo
a = 1
b = 2
n = 10000  # número de retângulos (ponto médio)

dx = (b - a) / n
x_mid = a + (np.arange(n) + 0.5) * dx  # pontos médios
area_num = np.sum(x_mid**3) * dx

print("Integral numérica (retângulos - ponto médio):", area_num)
print("Número de retângulos usados:", n)
```

### 1) Integral numérica (método dos retângulos)

Vamos aproximar a área sob a curva de $f(x)=x^3$ no intervalo $[1,2]$ usando o método dos retângulos (ponto médio).

Definimos $a=1$, $b=2$ e $n=10000$ retângulos.

$$\Delta x = \frac{b-a}{n} = \frac{2-1}{n}$$

Os pontos médios são:

$$x_i = a + \left(i+\frac{1}{2}\right)\Delta x,\quad i=0,1,\dots,n-1$$

A área aproximada é:

$$A \approx \sum_{i=0}^{n-1} f(x_i)\,\Delta x$$



```python
area_analitica = (2**4)/4 - (1**4)/4
print("Integral analítica (exata):", area_analitica)
```
